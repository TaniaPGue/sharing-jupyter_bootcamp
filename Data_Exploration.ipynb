# Data cleaning

```python
gapminder_copy = gapminder.copy()
```

```python
gapminder_copy.head(5)
```

```python
#Dropping NaN's 
gapminder_copy = gapminder_copy.dropna()
gapminder_copy.head(5)
```

```python
#Dropping duplicates
gapminder_copy = gapminder_copy.drop_duplicates(keep = "first")
```

```python
gapminder_copy.head(5)
```

```python
gapminder_copy.info()
```

```python
#Changing data types
gapminder_copy["year"] = gapminder_copy["year"].astype(int)
gapminder_copy["pop"] = gapminder_copy["pop"].astype(int)
```

```python
gapminder_copy.info()
```

```python
#Resetting the indices 
gapminder_copy = gapminder_copy.reset_index(drop = True)
gapminder_copy
```

```python
#Removing white spaces, 
gapminder_copy["region"] = gapminder_copy["region"].str.lstrip() #left justify 
gapminder_copy["region"] = gapminder_copy["region"].str.rstrip() #right justify
gapminder_copy["region"] = gapminder_copy["region"].str.lower() #all to lowercases 
```

```python
gapminder_copy
```

```python
gapminder_copy["country"] = gapminder_copy["region"].str.split("_", 1).str[1] 
#the first 1 stands for column not row. 
#the last part it has to do with where are you feeding this into this is making a new list

gapminder_copy["continent"] = gapminder_copy["region"].str.split("_", 1).str[0]

gapminder_copy.head()
```

```python
gapminder_copy = gapminder_copy.drop("region", 1)
gapminder_copy.head()
```

```python
#Renaming to lowercase
gapminder_copy.columns = gapminder_copy.columns.str.lower()
gapminder_copy.head()
```

```python
#Remove white space on strings using dictionaries 
gapminder_copy = gapminder_copy.rename(columns = {"life exp" : "lifeexp"})
gapminder_copy.head()
```

```python
PRB = pd.read_table("PRB_data.txt", sep = "\t")
```

```python
PRB.head()
```

```python
PRB["country"] = PRB["region"].str.split("_", 1).str[1].str.lower()
PRB["continent"] = PRB["region"].str.split("_", 1).str[0].str.lower()
PRB = PRB.drop("region", 1)
PRB.head()
```

```python
PRB.head()
```

```python
PRB.columns = PRB.columns.str.lower()
PRB = PRB.rename(columns = {"life exp" : "lifeexp"})
PRB.head()
```

```python
gapminder_comb = pd.concat([gapminder_copy, PRB])
gapminder_comb.tail()
```

```python
gapminder_comb = gapminder_comb.reset_index(drop = True)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
#so that Jupyper will automatically print the output 
import seaborn as sns 
```

```python
#Histogram
plt.hist(gapminder_copy["lifeexp"])
plt.xlabel("Life Expectancy")
plt.ylabel("Count")
```

```python
#Boxplot 
sns.boxplot(x = "year", y = "lifeexp", data = gapminder_copy)
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
```

```python
# Scattterplot

plt.scatter(gapminder_copy["gdppercap"], gapminder_copy["lifeexp"])
plt.xlabel("GDP per capita")
plt.ylabel("Life Expectancy")
```

```python
# scaling to log scale
plt.scatter(gapminder_copy["gdppercap"], gapminder_copy["lifeexp"])
plt.xlabel("GDP per capita (log)")
plt.ylabel("Life Expectancy")
plt.xscale("log")

plt.savefig("log-scale-scatter.png")
```
